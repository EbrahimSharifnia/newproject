```python
# Bootstrapping_Nuclear_Complete

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
from scipy.stats import norm
from scipy.stats import bootstrap
```

```python
df = pd.read_excel("Original_Data.xlsx")
#                    , sheet_name = "Pre-Bootstrap_Complete")
# df

df = df.drop(["Dates"], axis = 1)
heads = df.columns.to_list()
# heads
```

```python
df = df.fillna(df2)
```

```python
bootstrap_dic = {}

for i in heads:
    bootstrap_dic[i] = df[i].to_list()

# bootstrap_dic.values()
```

```python
bootstrap_dic_cleaned = {}
for i in heads:
    bootstrap_dic_cleaned[i] = [x for x in bootstrap_dic[i] if x == x]

# bootstrap_dic_cleaned
```

```python
data_mean_dic = {}
for i in heads:
    data_mean_dic[i] = np.nanmean(bootstrap_dic_cleaned[i])
    
data_std_dic = {}
for i in heads:
    data_std_dic[i] = np.nanstd(bootstrap_dic_cleaned[i])
```

```python
bootstrap_2_dic = {}
bootstrap_data_dic = {}
rng = np.random.default_rng()

for i in heads:
    if data_std_dic[i] == 0:
        dist = norm(loc=data_mean_dic[i] , scale=data_mean_dic[i]/24)
    else:
        dist = norm(loc=data_mean_dic[i] , scale=data_std_dic[i])
    data = dist.rvs(size=24, random_state=rng)
    bootstrap_data_dic[i] = data
    data = (data,)
    bootstrap_2_dic[i] = bootstrap(data, np.mean, confidence_level=0.95, random_state=rng)
```

```python
for i in heads:
    print(bootstrap_data_dic[i])
```

```python
bootstrap_df = pd.DataFrame(columns = heads)
# bootstrap_df
```

```python
for i in heads:
    bootstrap_df[i] = bootstrap_data_dic[i]
    
# bootstrap_df
```
