Splitting Data for Training and Testing

Autoregressive Moving Average (ARMA)

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
df_pred_last60 = pd.DataFrame(columns = heads)
for i in heads:
    y = df2[df2.index < pd.to_datetime("2019-07-01", format='%Y-%m-%d')][i].astype(float)
    ARMAmodel = SARIMAX(y, order = (1, 2, 1))
    ARMAmodel = ARMAmodel.fit()
    y_pred = ARMAmodel.get_forecast(12)
    y_pred_df = y_pred.conf_int(alpha = 0.05)
    y_pred_df["Predictions"] = ARMAmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1])
    # y_pred_df.index = test.index
    y_pred_out = y_pred_df["Predictions"]
    df_pred_last60[i] = y_pred_out
```
